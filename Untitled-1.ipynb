## 风速与扩散浓度的单项回归公式
 <img src="./docs/wind%20speed%20and%20concentraion.png" width = "600" height = "600" alt="风速与扩散浓度的单项回归公式" align=center />

```python
wind_v =  3.5
x_m = np.arange(0.000, 4.100, 0.030)
y_m = np.arange(0.000, 4.100, 0.030)
x_m, y_m = np.meshgrid(x_m, y_m)
concentration = 1.17223 + 0.00016*wind_v**3 - 0.00086*wind_v**2 - 0.00259*wind_v + 0.00385*x_m**3 - 0.016395*x_m**2 + 0.01175*x_m + 0.000493*y_m**3 - 0.0052*y_m**2 + 0.01344*y_m

# Creating figures for the plot
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('X-axis Distance', fontweight ='bold')  
ax.set_ylabel('Y-axis Distance', fontweight ='bold')  
ax.set_zlabel('Concentraion', fontweight ='bold')
def update(i):
    ax.view_init(i, 90)
plt.title("3D plot")

ax.plot_surface(x_m, y_m, concentration, rstride=1,cstride=1,cmap = 'rainbow')
anim = animation.FuncAnimation(
    fig,
    update,
    np.arange(0, 92), interval=10, blit=False
)
# with open('./anim.html', 'w', encoding='utf8') as file:
#     file.writelines(anim.to_jshtml(fps=25,embed_frames=True))
anim.save('anim.gif', writer='pillow',fps=25)
plt.show()
```
